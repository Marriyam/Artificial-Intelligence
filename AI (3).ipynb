```python
import pandas as pd
import matplotlib.pyplot as plt

dataset = pd.read_csv("Desktop/covid.csv")
df = pd.DataFrame(dataset, columns = ['Country', 'Total_Cases', 
                                    'Total_Deaths', 'Total_Recovered', 'Population'] )
  
# create histogram for numeric data
df.hist()

# show plot
plt.show()
```

```python

df.plot.bar()
  
# plot between 2 attributes
plt.bar(df['Total_Cases'], df['Total_Deaths'])
plt.xlabel("Total_Cases")
plt.ylabel("Total_Deaths")
plt.show()
```

```python
df.plot.box()
  
# individual attribute box plot
plt.boxplot(df['Total_Cases'])
plt.show()
```

```python
plt.pie(df['Total_Cases'], labels = {"A", "B", "C",
                             "D", "E", "F",
                             "G", "H", "I", "J"},
                               
autopct ='% 1.1f %%', shadow = True)
plt.show()
  
plt.pie(df['Total_Deaths'], labels = {"A", "B", "C",
                                "D", "E", "F",
                                "G", "H", "I", "J"},
                                  
autopct ='% 1.1f %%', shadow = True)
plt.show()
  
plt.pie(df['Total_Recovered'], labels = {"A", "B", "C",
                               "D", "E", "F",
                               "G", "H", "I", "J"},
autopct ='% 1.1f %%', shadow = True)
plt.show()
```

```python
# scatter plot between income and age
plt.scatter(df['Total_Cases'], df['Total_Deaths'])
plt.show()
  
# scatter plot between income and sales
plt.scatter(df['Total_Cases'], df['Total_Recovered'])
plt.show()
  
# scatter plot between sales and age
plt.scatter(df['Total_Recovered'], df['Total_Deaths'])
plt.show()
```
